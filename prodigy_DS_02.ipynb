```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
print("Initial missing values:\n", df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
print("\nAfter cleaning:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(16, 12))
```

```python
plt.subplot(2, 3, 1)
sns.countplot(data=df, x='Survived', palette='pastel')
plt.title('Survival Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
```

```python
plt.subplot(2, 3, 2)
sns.countplot(data=df, x='Sex', hue='Survived', palette='pastel')
plt.title('Survival by Gender')
```

```python
plt.subplot(2, 3, 3)
sns.countplot(data=df, x='Pclass', hue='Survived', palette='pastel')
plt.title('Survival by Passenger Class')
```

```python
plt.subplot(2, 3, 4)
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30, palette='pastel')
plt.title('Age Distribution by Survival')
```

```python
plt.subplot(2, 3, 5)
sns.countplot(data=df, x='SibSp', hue='Survived', palette='pastel')
plt.title('Survival by Siblings/Spouses Aboard')
```

```python
plt.subplot(2, 3, 6)
sns.countplot(data=df, x='Embarked', hue='Survived', palette='pastel')
plt.title('Survival by Embarkation Port')

plt.tight_layout()
plt.show()
```
